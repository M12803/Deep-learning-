```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

X = [1, 2, 3]

Wf, Uf, bf = 0.5, 0.1, 0
Wi, Ui, bi = 0.6, 0.2, 0
Wc, Uc, bc = 0.7, 0.3, 0
Wo, Uo, bo = 0.8, 0.4, 0

h = 0
C = 0

print("Initial hidden state:", h)
print("Initial cell state:", C)
print()


for t, x in enumerate(X, start=1):

    # Forget gate
    f = sigmoid(Wf * x + Uf * h + bf)

    # Input gate
    i = sigmoid(Wi * x + Ui * h + bi)

    # Candidate cell state
    C_tilde = tanh(Wc * x + Uc * h + bc)

    # Cell state update
    C = f * C + i * C_tilde

    # Output gate
    o = sigmoid(Wo * x + Uo * h + bo)

    # Hidden state update
    h = o * tanh(C)

    print(f"Time Step {t}")
    print("Input:", x)
    print("Forget gate:", round(f,3))
    print("Input gate:", round(i,3))
    print("Candidate state:", round(C_tilde,3))
    print("Cell state:", round(C,3))
    print("Output gate:", round(o,3))
    print("Hidden state:", round(h,3))
    #print("-"*40)


# Linear mapping from hidden state
Wy = 4.0
by = 0

y_pred = Wy * h + by

print("\nPredicted next value:", round(y_pred,3))
```
